```python
import pandas as pd
import praw
```

```python
# list of quarantined subreddits
quarantined_subs = pd.read_csv("quarantined-subs.txt", names = ['subreddit'])
```

```python
#setup the Reddit bot
praw_creds = pd.read_csv("praw-creds.txt")

reddit= praw.Reddit(client_id=praw_creds['client_id'][0],
                    client_secret=praw_creds['client_secret'][0],
                    password=praw_creds['password'][0],
                    user_agent="examining quarantined subreddits",
                    username=praw_creds['username'][0]) #Your reddit account information
```

```python
hits = 0

for sub in quarantined_subs['subreddit']:
    print(sub)
    subreddit = reddit.subreddit(sub)
    try:
        print(subreddit.subscribers)
        hits += 1
    except Exception as e:
        print(e)
```

```python
print("Number of non-quarantined, non-banned subreddits = ", hits)
```
